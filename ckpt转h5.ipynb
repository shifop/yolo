```python
import tensorflow as tf
import h5py
```

```python
reader = tf.train.NewCheckpointReader('./checkpoint/yolov4.ckpt')
with h5py.File('./checkpoint/tf_model.h5', 'w') as f:
    t_g = None
    nts = reader.get_variable_to_shape_map()
    for key in sorted(nts):
        key_ = key.replace('/','|').replace('Downsample|','')
        if 'yolov3|yolo-v3|Conv' in key_:
            continue
            
        key_ = key_.replace('Conv','conv2d').replace('weights','kernel')
        key_c = key_.split('|')
        if 'conv2d' in key_c[0] and 'BatchNorm'==key_c[1]:
            key_ = '|'.join([key_c[0].replace('conv2d','batch_normalization')]+key_c[2:])
        print('%s: %s'%(key_, nts[key]))
        f[key_] = reader.get_tensor(key)
```
